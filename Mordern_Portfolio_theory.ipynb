```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import scipy.stats 
import seaborn as sns 
```

```python
np.random.seed(51)

stocks = pd.DataFrame({"Stock_A": np.random.normal(10,1,size=10), "Stock_B": np.random.normal(10,5,size=10)})
stocks.index.name = "Months"
stocks = round(stocks,2)
stocks
```

```python
stocks["Stock_A_Rets"] =  stocks["Stock_A"] / stocks["Stock_A"].shift(1) - 1 

stocks["Stock_B_Rets"] = stocks["Stock_B"] / stocks["Stock_B"].shift(1) - 1

stocks = round(stocks, 2)
 
stocks
```
